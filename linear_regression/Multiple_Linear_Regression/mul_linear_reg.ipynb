```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('50_Startups.csv',index_col=0)
df
```

```python
############ encoder using label encoder from sklearn library #############
le = LabelEncoder()

df['State'] = le.fit_transform(df['State'])
df = df.reset_index()
```

```python
df.columns
```

```python
X = df.drop(['Profit'], axis=1)
```

```python
y = df[['Profit']]
```

```python
X = X.to_numpy()
X
```

```python
y = y.to_numpy()
y
```

```python
ones = np.ones((X.shape[0],1))
ones
Xbar = np.concatenate((ones,X),axis=1)
Xbar
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xbar,y,train_size=0.8, test_size=0.2, random_state=42)
```

```python
A = np.dot(X_train.T,X_train)
A
b = np.dot(X_train.T,y_train)
b
```

```python
w = np.dot(np.linalg.pinv(A),b)
w
```

```python
def calculate(prediction, actual):
    SSre = np.sum((actual - prediction)**2)
    SSmean = np.sum((actual - np.mean(actual))**2)
    R2_score = 1 - SSre/SSmean
    return R2_score

prediction = np.dot(X_test,w)
prediction
```

```python
R2_score = calculate(prediction,y_test)
R2_score
```

```python
######## Compare with library
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X_train,y_train)

y_prediction = lr.predict(X_test)
y_prediction
```

```python
from sklearn.metrics import r2_score

score = r2_score(y_test,y_prediction)
score
```
