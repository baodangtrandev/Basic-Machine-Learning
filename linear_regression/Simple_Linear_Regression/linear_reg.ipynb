```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183]]).T

Y = np.array([[49, 50, 51,  54, 58, 59, 60, 62, 63, 64, 66, 67, 68]]).T

print("X: ", X)
print("Y: ", Y)
```

```python
plt.plot(X,Y,'bo')
ax = plt.axis([140,190,45,75])
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
one = np.ones((X.shape[0],1))
one

Xbar = np.concatenate((one,X), axis=1)
Xbar
```

```python
A = np.dot(Xbar.T,Xbar)
b = np.dot(Xbar.T,Y)

w = np.dot(np.linalg.pinv(A),b)

w
```

```python
w0 = w[0][0]

w1 = w[1][0]

x0 = np.linspace(140,190,2)

y0 = w0 + np.dot(x0,w1)

y0
```

```python
plt.plot(x0,y0,'r')
plt.plot(X,Y,'bo')
plt.axis([140,190,45,75])
plt.title("Linear Regression")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.show()
```

```python
y1 = w0 + 155*w1
y2 = w0 + 160*w1

print( u'Predict weight of person with height 155 cm: %.2f (kg)'  %(y1) )
print( u'Predict weight of person with height 160 cm: %.2f (kg)'  %(y2) )
```

```python
######### Compare with linear regression in scikitlearn

from sklearn import linear_model, datasets

regr = linear_model.LinearRegression(fit_intercept=False)
regr.fit(Xbar, Y)


# Compare two results
print( 'Solution found by scikit-learn  : ', regr.coef_ )
print( 'Solution found by (5): ', w.T)
```
